```python
import os
import shutil
import random
from google.colab import drive

# Step 1: Mount Google Drive
drive.mount('/content/drive')

# Step 2: Define dataset path
dataset_path = "/content/LicenseDetector/Train"  # Update this path
train_path = "/content/drive/MyDrive/Dataset/Train"
test_path ="/content/drive/MyDrive/Dataset/Test"

# Step 3: Create train and test directories if they don't exist
os.makedirs(train_path, exist_ok=True)
os.makedirs(test_path, exist_ok=True)

# Step 4: Get all image filenames
all_images = [f for f in os.listdir(dataset_path) if f.endswith('.png')]

# Step 5: Shuffle and split (80% training, 20% testing)
random.seed(42)  # Ensure reproducibility
random.shuffle(all_images)
split_idx = int(0.8 * len(all_images))

train_images = all_images[:split_idx]
test_images = all_images[split_idx:]

# Step 6: Move images to respective directories
for img in train_images:
    shutil.move(os.path.join(dataset_path, img), os.path.join(train_path, img))

for img in test_images:
    shutil.move(os.path.join(dataset_path, img), os.path.join(test_path, img))

print(f"Total images: {len(all_images)}")
print(f"Training images: {len(train_images)}")
print(f"Testing images: {len(test_images)}")
```

```python
import os
import shutil
from PIL import Image

# Define the folder where images are stored
folder_path = "/content/drive/MyDrive/LicencePlateDetector/Train2"

# Convert all images to PNG format
for filename in os.listdir(folder_path):
    if filename.lower().endswith(("jpg", "jpeg", "bmp", "gif", "tiff", "webp")):
        file_path = os.path.join(folder_path, filename)
        img = Image.open(file_path)

        # Convert to PNG and overwrite the original file
        new_file_path = os.path.join(folder_path, os.path.splitext(filename)[0] + ".png")
        img.save(new_file_path, "PNG")

        # Remove the old file
        os.remove(file_path)
        print(f"Converted and replaced: {filename} -> {new_file_path}")


# Define the source and temporary writable directory
source_folder = "/content/drive/MyDrive/LicencePlateDetector/Dataset/Test"
temp_folder = "/content/temp_test"

# Create temp folder if it doesn't exist
os.makedirs(temp_folder, exist_ok=True)

# Copy all images to temp folder (to allow modifications)
for filename in os.listdir(source_folder):
    if filename.lower().endswith(("jpg", "jpeg", "bmp", "gif", "tiff", "webp")):
        shutil.copy(os.path.join(source_folder, filename), temp_folder)

# Convert all images to PNG format in the temp folder
for filename in os.listdir(temp_folder):
    if filename.lower().endswith(("jpg", "jpeg", "bmp", "gif", "tiff", "webp")):
        file_path = os.path.join(temp_folder, filename)
        img = Image.open(file_path)

        # Convert to PNG
        new_file_path = os.path.join(temp_folder, os.path.splitext(filename)[0] + ".png")
        img.save(new_file_path, "PNG")

        # Remove old file
        os.remove(file_path)
        print(f"Converted: {filename} -> {new_file_path}")

destination_folder = "/content/drive/MyDrive/LicencePlateDetector/Test2"

# Move all files
shutil.move(temp_folder, destination_folder)

print("All files moved to NewTest folder in Google Drive.")
print(" All test files converted to PNG and moved back to Google Drive.")
```

```python
IMAGE_SIZE = [224, 224]
train_path=' /content/drive/MyDrive/LicencePlateDetector/Train2'
valid_path=' /content/drive/MyDrive/LicencePlateDetector/Test2'
```

```python
inception= InceptionV3(input_shape=IMAGE_SIZE+[3], weights= 'imagenet', include_top=False)
```

```python
for layer in inception.layers:
  layer.trainable=False
```

```python
folders =glob('/content/drive/MyDrive/LicencePlateDetector/Train2')
```

```python
folders
```

```python
x=Flatten()(inception.output)
```

```python
prediction=Dense(len(folders), activation='softmax')(x)
```

```python
model=Model(inputs=inception.input, outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
#data augmentation
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
training_set=train_datagen.flow_from_directory('/content/drive/MyDrive/LicencePlateDetector/Train2',
                                               target_size=(224,224),
                                               batch_size=16,
                                               class_mode='categorical')
```

```python
test_set=test_datagen.flow_from_directory('/content/drive/MyDrive/LicencePlateDetector/Test2',
                                               target_size=(224,224),
                                               batch_size=16,
                                               class_mode='categorical')
```
